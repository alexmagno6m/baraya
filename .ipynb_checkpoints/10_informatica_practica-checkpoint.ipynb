1. En ```Visual Studio Code``` cree un nuevo folder llamado ```nivelacion_nombre_apellido_curso```, dentro de ese folder agregue los siguientes archivos y directorios como se muestra en este ejemplo:

```
└── 📁nivelacion_pedro_perez_1002/
    ├── 📁CSS/
    │   └── 🗎 style.css
    └── 🗎 pedro_perez_curso_index.html
```

3. Escriba como titulo de la pagina: "Nivelacion grado 10 CSS y HTML".
4. Como titulo ```H1``` escriba "Certificacion de HTML y CSS"
5. Agregue una linea horizontal con ```hr```.
6. Cree la estructura de ```div``` similar a la siguiente representacion

7. Agregue en los ```div``` hijos la siguiente informacion respectivamente, cada una de las cuales debe ir dentro de una etiqueta ```p```:
> - HTML (HyperText Markup Language): HTML es el lenguaje de marcado utilizado para estructurar y organizar el contenido en la web. Utiliza etiquetas para definir elementos como encabezados, párrafos, enlaces, imágenes y otros elementos que constituyen una página web. HTML proporciona la estructura básica sobre la cual se construyen las páginas web, permitiendo a los navegadores interpretar y renderizar el contenido de manera coherente
> - CSS (Cascading Style Sheets): CSS es un lenguaje de estilo utilizado para dar formato y diseño a las páginas web creadas con HTML. Permite controlar la presentación visual de los elementos HTML, definiendo propiedades como colores, fuentes, márgenes, tamaños y disposición en la página. CSS facilita la separación de la estructura del contenido (definida en HTML) y su presentación visual, lo que permite un desarrollo web más modular y mantenible.
> - JavaScript: JavaScript es un lenguaje de programación que se utiliza para agregar interactividad y dinamismo a las páginas web. A diferencia de HTML y CSS, que se centran en la estructura y el diseño, JavaScript permite la manipulación del contenido en tiempo real, la interacción con el usuario, la validación de formularios y la actualización dinámica de la página sin necesidad de recargarla. Es un componente esencial para el desarrollo de aplicaciones web modernas y proporciona la capacidad de crear experiencias interactivas y atractivas en el navegador del usuario.

8. Agregue en el archivo ```style```, los estilos correspondientes para:
